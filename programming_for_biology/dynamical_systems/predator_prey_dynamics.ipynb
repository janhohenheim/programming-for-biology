```python
from programming_for_biology.for_loops.population_dynamics import Population, get_population_count_hourly
import numpy as np
import matplotlib.pyplot as plt
```

```python
initial_population = Population(rabbit_count=1000, fox_count=100)
days = np.arange(0, 200, 1)
populations = [get_population_count_hourly(initial_population, day) for day in days]
foxes = [population.fox_count for population in populations]
rabbits = [population.rabbit_count for population in populations]
plt.title("Population over time")
plt.plot(days, foxes, label="Foxes")
plt.plot(days, rabbits, label="Rabbits")
plt.xlabel("Days")
plt.ylabel("Population")
plt.grid()
plt.legend()
plt.show()
print("Maximal fox count:", max(foxes))
print("Maximal rabbit count:", max(rabbits))
```
